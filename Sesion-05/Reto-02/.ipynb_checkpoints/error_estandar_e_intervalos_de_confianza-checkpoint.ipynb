```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#analizaremos el dataset de objetos cercanos a la tierra
df = pd.read_csv('../../Datasets/melbourne_housing-clean.csv', index_col=0)

df.head()
```

```python
#Analizamos el datafram,nos finamos en la variable precio
df.describe()
```

```python
muestra=df['price']
```

```python
#Distrubucion muestral de la media
means = []
for i in range(100_000):
    sample = muestra.sample(n=50, replace=True)
    means.append(sample.mean())
    
serie_means = pd.Series(means)
```

```python
#Distrubucion muestral de la mediana

medians = []
for i in range(100_000):
    sample = muestra.sample(n=len(muestra), replace=True)
    medians.append(sample.median())
    
serie_medians = pd.Series(medians)
```

```python
#Grafica distribucion muestral media
sns.distplot(serie_means, kde=False, norm_hist=False);
```

```python
#Grafica distribucion muestral mediana
sns.distplot(serie_medians, kde=False, norm_hist=False);
```

```python
#error estandar
print(f'Error estandar: {serie_means.std()}')
print(f'Valor mínimo: {serie_means.min()}')
print(f'Valor máximo: {serie_means.max()}')
print(f'Rango: {serie_means.max() - serie_means.min()}')
```

```python
print(f'Error estandar: {serie_medians.std()}')
print(f'Valor mínimo: {serie_medians.min()}')
print(f'Valor máximo: {serie_medians.max()}')
print(f'Rango: {serie_medians.max() - serie_medians.min()}')
```
